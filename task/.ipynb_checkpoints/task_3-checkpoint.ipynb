```python
with  open( 'input.txt', 'r') as file:
    a,*b = file.read().split()
    a=int(a)
    b=[ list(x) for x in b]
    

# check I
def tablo():
    i = [tuple([ y for y, x in enumerate (b[i]) if x=='#']) for i in range(0,a)] # список индексов #
    i_1 = [ x for x in i if x] # удалили  пустые  строки

    if not len(i_1):
        return print('X')

    i_ind = [ x  for x,y in   enumerate(i) if y] # список индексов  строк с #
    i_o=set(i_1) # уникальные значения строк там где есть #
    i_ind_cond =all( [i_ind[x]+1 == i_ind[x+1] for x in range(0, len(i_ind)-1)]) # проверка непрерывность  строк с #

    # проверка  на непрерывность длинной строки
    i_o_max =max(i_o, key=len)
    i_o_ind_conts_h= all( [i_o_max[x]+1 == i_o_max[x+1] for x in range(0, len(i_o_max)-1)])
    
    # print(i,i_1,i_ind, i_ind_cond, i_o_ind_conts_h)
    if not i_ind_cond or len(i_o) > 3 or len(i_o)==0 or not i_o_ind_conts_h:
        return print ('X')
    
    if len(set(tuple(i_1)))== 1: #  уникальность  строки с # + проверка на непрерывность строки #
        return print('I')
 
    # Word O
    if len(i_o)==2:
         i_o_ind = [x for x,y in enumerate(i) if y== min(i_o, key=len)] #  индексы строк с # (короткие)  
         i_o_ind_long =[x for x,y in enumerate(i) if y== max(i_o, key=len)] #  индексы строк с # (длинные) 
         insert_o = max( i_o_ind_long)> max(i_o_ind) and min(i_o_ind_long ) < min (i_o_ind)
        
          #  проверка короткой строки на внутренний пропуск пропуск 
         o_1, o_2 = i_o
         o_2_skip = [*set(o_1)^set(o_2)]
         o_short_cond =all( [o_2_skip[x]+1 == o_2_skip[x+1] for x in range(0, len(o_2_skip)-1)])
          
          # проверка на последовательность индексов внутренних коротких строк
         i_o_ind_conts= all( [i_o_ind[x]+1 == i_o_ind[x+1] for x in range(0, len(i_o_ind)-1)])

          # проверка  замкнутости прямоугольника, есть смысл  при выполнении условия
         cloused_o =  o_1[0] ==o_2[0] and o_1[len(o_1)-1] ==o_2[len(o_2)-1]
         if i_o_ind_conts and  insert_o and i_ind_cond and cloused_o and o_short_cond:
             return print ('O')


         # word C
        
    if len(i_o) == 2:
             
             #проверка короткой строки
        o_2 = min(i_o, key=len)
        c_short_cond =all( [o_2[x]+1 == o_2[x+1] for x in range(0, len(o_2)-1)])

             # проверка  замкнутости прямоугольника, есть смысл  при выполнении условия
        o_1, o_2 = i_o
        cloused_c =  o_1[0] ==o_2[0]  and o_1[len(o_1)-1] !=o_2[len(o_2)-1]

        if i_ind_cond and insert_o and i_o_ind_conts and cloused_c and  c_short_cond:
            return print ('C')


          # word L
    if len(i_o) == 2:
              # Вложенность строк для L ( короткие выше длинных)
        insert_l = min( i_o_ind_long) > max(i_o_ind)  

        if insert_l and i_o_ind_conts and cloused_c and c_short_cond:
            return print ('L')


       # word H
    if len(i_o)==2:

        # блок проверки  размещениия коротких и длинных строк 
        insert_h = max( i_o_ind_long)< max(i_o_ind) and min(i_o_ind_long ) > min (i_o_ind)
        
        i_o_ind_conts_h= all( [i_o_ind_long[x]+1 == i_o_ind_long[x+1] for x in range(0, len(i_o_ind_long)-1)]) 
        
        o_1, o_2 = i_o
        cloused_h =  o_1[0] ==o_2[0]  and o_1[len(o_1)-1] ==o_2[len(o_2)-1]

        if i_ind_cond and insert_h and i_o_ind_conts_h and cloused_h and o_short_cond:
             return print ('H')


       # word P
    if len(i_o) == 3:
    
        # список строк, отсортиорованный по количеству #
        i_p = sorted(i_o, key=len)

        i_p_ind = [x for x,y in enumerate(i) if y== i_p[0]] #  индексы строк с # (короткие)  
        i_p_ind_long =[x for x,y in enumerate(i) if y== i_p[2]] #  индексы строк с # (длинные) 
        i_p_ind_middle=[x for x,y in enumerate(i) if y== i_p[1]] #  индексы строк с # (средние)
     

        #проверка коротких  строк  на  внутренний пропуск
    # для короткой строки
        p_2_skip = [*set(i_p[0])^set(i_p[2])]
        p_short_cond_1 =all( [p_2_skip[x]+1 == p_2_skip[x+1] for x in range(0, len(p_2_skip)-1)])

     # для средней строки
        p_1_skip = [*set(i_p[1])^set(i_p[2])]
        p_short_cond_2 =all( [p_1_skip[x]+1 == p_1_skip[x+1] for x in range(0, len(p_1_skip)-1)])

        p_short_cond = p_short_cond_1 and p_short_cond_2


        insert_p_1 = min( i_p_ind)> max(i_p_ind_long) 
        insert_p_2 = min( i_p_ind_middle)>min(i_p_ind_long) and max( i_p_ind_middle) < max(i_p_ind_long)

     # проверка на последовательность индексов средних внутренних строк
        i_p_ind_conts= all( [i_p_ind_middle[x]+1 == i_p_ind_middle[x+1] for x in range(0, len(i_p_ind_middle)-1)])

    # проверка  замкнутости прямоугольника, есть смысл  при выполнении условия
        cloused_p_1 = i_p[0][0] == i_p[1][0] == i_p[2][0]
        cloused_p_2 = i_p[1][-1] == i_p[2][-1]

        cloused_4 = i_p[0]== i_p[1][: len(i_p[0])] # равенство короткой строки  и начала  средней
        if len (i_p[1]) >len (i_p[0]):
            cloused_5 = i_p[1][len(i_p[0])-1]+1 !=i_p[1][len(i_p[0])] # проверка  что следующий средней строки  элемент - пропуск
        else: cloused_5 =False # проверка  что следующий средней строки  элемент - пропуск
    
        cloused_p =  cloused_p_1 and cloused_p_2 and cloused_4 and cloused_5


        if  i_p_ind_conts and  p_short_cond and i_o_ind_conts_h and cloused_p and insert_p_1 and insert_p_2 :
            return print ('P')
    return print ('X')
tablo()
```
